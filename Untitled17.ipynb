```python
class Mahasiswa:
    def __init__(self, nama, nim, jurusan):
        """Konstruktor untuk inisialisasi atribut mahasiswa."""
        self.nama = nama
        self.nim = nim
        self.jurusan = jurusan


    def tampilkan_biodata(self):
        """Metode untuk menampilkan biodata mahasiswa."""
        print("\nBiodata Mahasiswa")
        print("-----------------")
        print(f"Nama      : {self.nama}")
        print(f"NIM       : {self.nim}")
        print(f"Jurusan   : {self.jurusan}")


# Fungsi utama untuk input data mahasiswa
def main():
    print("Masukkan data mahasiswa")
    nama = input("Nama      : ")
    nim = input("NIM       : ")
    jurusan = input("Jurusan   : ")


    # Membuat objek mahasiswa
    mahasiswa = Mahasiswa(nama, nim, jurusan)

    # Menampilkan biodata mahasiswa
    mahasiswa.tampilkan_biodata()

# Memanggil fungsi utama
if __name__ == "__main__":
    main()
```

```python
class DataMahasiswa:
    def __init__(self):
        """Inisialisasi atribut default."""
        self.__nama = ""
        self.__nim = ""
        self.__jurusan = ""

    # Getter dan Setter untuk Nama
    def get_nama(self):
        return self.__nama

    def set_nama(self, nama):
        if len(nama) > 0:
            self.__nama = nama
        else:
            print("Nama tidak boleh kosong!")

    # Getter dan Setter untuk NIM
    def get_nim(self):
        return self.__nim

    def set_nim(self, nim):
        if nim.isdigit() and len(nim) == 12:
            self.__nim = nim
        else:
            print("NIM harus berupa angka 12 digit!")

    # Getter dan Setter untuk Jurusan
    def get_jurusan(self):
        return self.__jurusan

    def set_jurusan(self, jurusan):
        if jurusan in ["Informatika", "Sistem Informasi", "Teknik Komputer"]:
            self.__jurusan = jurusan
        else:
            print("Jurusan tidak valid! Pilih salah satu: Informatika, Sistem Informasi, Teknik Komputer.")

    # Method untuk menampilkan data mahasiswa
    def tampilkan_biodata(self):
        print("\nBiodata Mahasiswa")
        print("-----------------")
        print(f"Nama      : {self.__nama}")
        print(f"NIM       : {self.__nim}")
        print(f"Jurusan   : {self.__jurusan}")


# Program utama
def main():
    mahasiswa = DataMahasiswa()

    while True:
        print("\n1. Input Data Mahasiswa")
        print("2. Ubah Data Mahasiswa")
        print("3. Tampilkan Biodata")
        print("4. Keluar")
        pilihan = input("Pilih menu (1/2/3/4): ")

        if pilihan == "1":
            nama = input("Masukkan Nama: ")
            mahasiswa.set_nama(nama)

            nim = input("Masukkan NIM (12 digit): ")
            mahasiswa.set_nim(nim)

            jurusan = input("Masukkan Jurusan (Informatika/Sistem Informasi/Teknik Komputer): ")
            mahasiswa.set_jurusan(jurusan)

        elif pilihan == "2":
            print("\n1. Ubah Nama")
            print("2. Ubah NIM")
            print("3. Ubah Jurusan")
            sub_pilihan = input("Pilih data yang ingin diubah (1/2/3): ")

            if sub_pilihan == "1":
                nama = input("Masukkan Nama baru: ")
                mahasiswa.set_nama(nama)

            elif sub_pilihan == "2":
                nim = input("Masukkan NIM baru (10 digit): ")
                mahasiswa.set_nim(nim)

            elif sub_pilihan == "3":
                jurusan = input("Masukkan Jurusan baru (Informatika/Sistem Informasi/Teknik Komputer): ")
                mahasiswa.set_jurusan(jurusan)

            else:
                print("Pilihan tidak valid!")

        elif pilihan == "3":
            mahasiswa.tampilkan_biodata()

        elif pilihan == "4":
            print("Terima kasih! Program selesai.")
            break

        else:
            print("Pilihan tidak valid! Silakan coba lagi.")

# Memanggil fungsi utama
if __name__ == "__main__":
    main()
```
